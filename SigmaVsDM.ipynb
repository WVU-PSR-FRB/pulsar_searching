```python
import numpy as np
import matplotlib.pyplot as plt
```

### Extracting range of DM vlaues

<font color = 'blue'> Below are the values taken from dedispersion file to created range of DM values used to dedisperse the file. We need those values to pick DM values while doing stacksearch so this block of code below will create an array of those DM values that can be used later.</font>

```python
dDMs = [0.1, 0.2, 0.3, 0.5, 0.5, 1.0]
startDMs = [0.0, 176.4,  319.20, 470.40, 764.40, 964.40,1014.4]

dm_array_values = [] # Will give us required range of DM values. 
for ind1 in np.arange(0, len(dDMs)):
        init_val = startDMs[ind1]
        final_val = startDMs[ind1+1]
        for ind2 in np.arange(init_val,final_val,dDMs[ind1]):
            dm_value = np.char.mod("%.2f", ind2)
            dm_array_values.append(dm_value)
```

### Extracting DM and highest sigma values

<font color = 'blue'>Code below will create a log of text file with sigma values and dm_values. Will go through all the output text files created as a result of stack search and pick highest sigma value and DM value of the file. By creating a graph for sigma vs dm we can figure out which DM value sticks out. </font>

```python
with open('sigma_logs.txt', "w") as outfile: 
    for n1 in np.arange(0, len(dDMs)):
        init_val = startDMs[n1]
        final_val = startDMs[n1+1]
        for n2 in np.arange(init_val,final_val,dDMs[n1]):
            dm_value = np.char.mod('%.2f', n2)
            file_name = "output_DM_"+ str(dm_value)+".txt"
            try:
                with open(file_name, "r") as infile:
            
                    for i, line in enumerate(infile, start=1):
                        if i==3:
                            sigma_part = line.split("    ")[0]
                            # text_to_write = "\n" + sigma_part + "\t" + str(dm_value)
                            outfile.write(f"{sigma_part} \t\t\t {str(dm_value)}\n")
                            # outfile.writelines(sigma_part)
                            # outfile.write("\t")
                            # outfile.writelines(str(dm_value))
                            break  
            except FileNotFoundError:
                print("Output file", file_name, "was not found.")
                
```

<font color = 'blue'>More or less code below is similar to above but the one below instead of creating a text file it gives DM values and sigma values in two different arrays which can then be used to make a graph and see what DM value sticks out. </font>

```python
dm_array = []
sigma_array = []
for n1 in np.arange(0, len(dDMs)):
    init_val = startDMs[n1]
    final_val = startDMs[n1+1]
    for n2 in np.arange(init_val,final_val,dDMs[n1]):
        dm_value = np.char.mod('%.2f', n2)
        file_name = "output_DM_"+ str(dm_value)+".txt"
        try:  
            with open(file_name, "r") as infile:
                for i, line in enumerate(infile, start=1):
                    if i==3:
                        sigma_part = line.split("    ")[0]
                        dm_array.append(float(dm_value))
                        sigma_array.append(float(sigma_part))
                        # print(dm_value, '\t\t', sigma_part)
                        break   
        except FileNotFoundError:
            print("Output file", file_name, "was not found.")
```

### Plot of Sigma Vs DM 
<font color = 'blue'>Create a plot of sigma value vs DM value to see if any particular DM value sticks out.</font>

```python
dm = dm_array
sigma = sigma_array
fig, sigmaVSdm = plt.subplots(figsize=(15,10), layout = 'constrained')
sigmaVSdm.plot(dm, sigma)
sigmaVSdm.set_xlabel('DM_values', fontweight = 'bold', fontsize = 14)
sigmaVSdm.set_ylabel('Sigma_values', fontweight = 'bold', fontsize = 14)
plt.savefig('Sigma Vs DM plot')
plt.close()
```
